### ImpIementing K Means

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import helpers2 as h
import tests as t
from IPython import display

%matplotlib inline

# Make the images larger
plt.rcParams['figure.figsize'] = (16, 9)
```

```python
data = h.simulate_data(200, 5, 4)# Creating a dataset with 200 points, 5 features and 4 centers
```

Steps to use KMeans:

**I.** Instantiate your model.

**II.** Fit your model to the data.

**III.** Predict the labels for the data.

```python
# Instantiating a model with 4 centers
kmeans_4 =KMeans(4) #instantiate your model

model_4 = kmeans_4.fit(data)#fitting the model to your data using kmeans_4

labels_4 = kmeans_4.predict(data)#predicting labels using model_4 on your dataset

# plotting data colored by center
h.plot_data(data, labels_4)
```

fitting kmeans using 2 clusters instead of 4

```python
kmeans_2 = KMeans(2)
model_2 = kmeans_2.fit(data)
labels_2 = model_2.predict(data)
h.plot_data(data, labels_2)
```

```python
kmeans_7 = KMeans(7)
model_7 = kmeans_7.fit(data)
labels_7 = model_7.predict(data)
h.plot_data(data, labels_7)
```

As you can see, each time additional centers are considered, the distances between the points and the center will decrease.  However, at some point, that decrease is not substantial enough to suggest the need for an additional cluster.  

Using a scree plot is a common method for understanding if an additional cluster center is needed.  The elbow method used by looking at a scree plot is still pretty subjective, but let's take a look to see how many cluster centers might be indicated.

```python
# Fitting a kmeans model with changing k from 1-10
# Obtain the score for each model (taking the absolute value)
# Plotting the score against k
```

```python
centers, scores = h.fit_mods()

plt.plot(centers, scores, linestyle='--', marker='o', color='b');
plt.xlabel('K');
plt.ylabel('SSE');
plt.title('SSE vs. K');
```

The scree plot, suggests 4 is appropriate K value as that is the elbow of curve.
